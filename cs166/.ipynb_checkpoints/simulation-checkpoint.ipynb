# CS166: Restroom Queue Simulation (Cellular Automata)

This notebook implements a **cellular automata** simulation where you can see:
- The grid layout with stalls, sinks, queues
- People (agents) moving on the grid
- Queues forming and people waiting
- Occupancy of fixtures

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.colors import ListedColormap
from matplotlib.animation import FuncAnimation
from IPython.display import HTML, display, clear_output
from dataclasses import dataclass, field
from typing import List, Optional, Tuple, Literal
from enum import IntEnum
import time
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

## 1. Cell Types and Colors

Define the grid cell types and their visual representation.


```python
# Cell types
class CellType(IntEnum):
    EMPTY = 0
    WALL = 1
    W_STALL = 2      # Women's stall
    M_STALL = 3      # Men's stall
    URINAL = 4
    SINK = 5
    QUEUE_W = 6      # Women's queue
    QUEUE_M = 7      # Men's queue
    QUEUE_SHARED = 8 # Shared queue
    ENTRANCE = 9
    EXIT = 10

# Colors for each cell type (RGB normalized 0-1)
CELL_COLORS = {
    CellType.EMPTY: '#f5f5f5',       # Light grey floor
    CellType.WALL: '#333333',         # Dark grey walls
    CellType.W_STALL: '#ffc1e3',      # Pink for women's stalls
    CellType.M_STALL: '#b3d9ff',      # Light blue for men's stalls
    CellType.URINAL: '#4a90e2',       # Darker blue for urinals
    CellType.SINK: '#ce93d8',         # Purple for sinks
    CellType.QUEUE_W: '#ffccf2',      # Light pink queue
    CellType.QUEUE_M: '#d6eaff',      # Light blue queue
    CellType.QUEUE_SHARED: '#e1f5fe', # Cyan for shared queue
    CellType.ENTRANCE: '#c8e6c9',     # Green entrance
    CellType.EXIT: '#ffccbc',         # Orange exit
}

# Person states
class PersonState(IntEnum):
    WALKING_TO_QUEUE = 0
    IN_QUEUE = 1
    WALKING_TO_STALL = 2
    IN_STALL = 3
    WALKING_TO_SINK = 4
    AT_SINK = 5
    EXITING = 6
    DONE = 7

STATE_NAMES = {
    PersonState.WALKING_TO_QUEUE: "‚ÜíQueue",
    PersonState.IN_QUEUE: "Waiting",
    PersonState.WALKING_TO_STALL: "‚ÜíStall",
    PersonState.IN_STALL: "Using",
    PersonState.WALKING_TO_SINK: "‚ÜíSink",
    PersonState.AT_SINK: "Washing",
    PersonState.EXITING: "Leaving",
    PersonState.DONE: "Done",
}
```

## 2. Core Classes: Fixtures and People


```python
@dataclass
class Stall:
    """A toilet stall or urinal fixture."""
    col: int
    row: int
    stall_type: str  # 'w_stall', 'm_stall', 'urinal'
    gender_allowed: str  # 'F', 'M', or 'both'
    entrance_col: int
    entrance_row: int
    occupant_id: Optional[int] = None
    occupied_until: float = 0.0

@dataclass
class Sink:
    """A sink fixture."""
    col: int
    row: int
    entrance_col: int
    entrance_row: int
    occupant_id: Optional[int] = None
    occupied_until: float = 0.0

@dataclass
class QueueCell:
    """A cell in the queue."""
    col: int
    row: int
    gender: str  # 'F', 'M', or 'shared'

@dataclass
class Person:
    """A person/agent in the simulation."""
    id: int
    col: int
    row: int
    gender: str  # 'F' or 'M'
    dwell_time: float  # Time to spend in stall
    sink_time: float   # Time to spend at sink
    time_entered: float  # When they entered the system
    state: PersonState = PersonState.WALKING_TO_QUEUE
    target_queue_idx: Optional[int] = None
    target_stall: Optional[Stall] = None
    target_sink: Optional[Sink] = None
    time_entered_stall: Optional[float] = None
    time_entered_sink: Optional[float] = None
```

## 3. The Grid Class

The grid represents the physical layout of the bathroom.


```python
class Grid:
    """The bathroom layout grid."""
    
    def __init__(self, cols: int, rows: int):
        self.cols = cols
        self.rows = rows
        self.cells = np.zeros((rows, cols), dtype=int)  # CellType values
        self.stalls: List[Stall] = []
        self.sinks: List[Sink] = []
        self.queue_cells_women: List[QueueCell] = []
        self.queue_cells_men: List[QueueCell] = []
        self.queue_cells_shared: List[QueueCell] = []
        self.entrance_cell: Optional[Tuple[int, int]] = None
        self.exit_cell: Optional[Tuple[int, int]] = None
        self.entrance_women: Optional[Tuple[int, int]] = None
        self.entrance_men: Optional[Tuple[int, int]] = None
    
    def set_cell(self, row: int, col: int, cell_type: CellType):
        if 0 <= row < self.rows and 0 <= col < self.cols:
            self.cells[row, col] = cell_type
    
    def get_cell(self, row: int, col: int) -> CellType:
        if 0 <= row < self.rows and 0 <= col < self.cols:
            return CellType(self.cells[row, col])
        return CellType.WALL
    
    def draw_border(self):
        self.cells[0, :] = CellType.WALL
        self.cells[-1, :] = CellType.WALL
        self.cells[:, 0] = CellType.WALL
        self.cells[:, -1] = CellType.WALL
    
    def add_stall(self, col: int, row: int, gender: str, stall_type: str = 'stall'):
        if stall_type == 'urinal':
            cell_type = CellType.URINAL
            fixture_type = 'urinal'
            gender_allowed = 'M'
        elif gender == 'F':
            cell_type = CellType.W_STALL
            fixture_type = 'w_stall'
            gender_allowed = 'F'
        else:
            cell_type = CellType.M_STALL
            fixture_type = 'm_stall'
            gender_allowed = 'M'
        
        self.set_cell(row, col, cell_type)
        self.stalls.append(Stall(
            col=col, row=row, stall_type=fixture_type,
            gender_allowed=gender_allowed,
            entrance_col=col, entrance_row=row + 1  # Entrance below
        ))
    
    def add_sink(self, col: int, row: int):
        self.set_cell(row, col, CellType.SINK)
        self.sinks.append(Sink(
            col=col, row=row,
            entrance_col=col, entrance_row=row + 1
        ))
    
    def add_queue_cell(self, col: int, row: int, gender: str):
        if gender == 'F':
            cell_type = CellType.QUEUE_W
            self.queue_cells_women.append(QueueCell(col, row, gender))
        elif gender == 'M':
            cell_type = CellType.QUEUE_M
            self.queue_cells_men.append(QueueCell(col, row, gender))
        else:
            cell_type = CellType.QUEUE_SHARED
            self.queue_cells_shared.append(QueueCell(col, row, 'shared'))
        self.set_cell(row, col, cell_type)
    
    def set_entrance(self, col: int, row: int, gender: Optional[str] = None):
        self.set_cell(row, col, CellType.ENTRANCE)
        if gender == 'F':
            self.entrance_women = (col, row)
        elif gender == 'M':
            self.entrance_men = (col, row)
        else:
            self.entrance_cell = (col, row)
    
    def set_exit(self, col: int, row: int):
        self.set_cell(row, col, CellType.EXIT)
        self.exit_cell = (col, row)
```

## 4. Pathfinding (BFS)

Simple BFS to find the next step toward a target.


```python
def find_next_step(start_col: int, start_row: int, goal_col: int, goal_row: int,
                   is_walkable, cols: int, rows: int) -> Optional[Tuple[int, int]]:
    """BFS pathfinding - returns the next cell to move to."""
    if start_col == goal_col and start_row == goal_row:
        return None
    
    from collections import deque
    
    queue = deque([(start_col, start_row, None)])  # (col, row, parent)
    visited = {(start_col, start_row)}
    parent_map = {}
    
    directions = [(0, -1), (1, 0), (0, 1), (-1, 0)]  # up, right, down, left
    
    while queue:
        col, row, parent = queue.popleft()
        
        if col == goal_col and row == goal_row:
            # Backtrack to find first step
            path = []
            current = (col, row)
            while current in parent_map:
                path.append(current)
                current = parent_map[current]
            if path:
                return path[-1]  # First step from start
            return None
        
        for dc, dr in directions:
            nc, nr = col + dc, row + dr
            if 0 <= nc < cols and 0 <= nr < rows and (nc, nr) not in visited:
                if is_walkable(nc, nr):
                    visited.add((nc, nr))
                    parent_map[(nc, nr)] = (col, row)
                    queue.append((nc, nr, (col, row)))
    
    return None  # No path found
```

## 5. The CA Simulation Engine

This is the main simulation that moves agents on the grid.


```python
class CASimulation:
    """Cellular Automata Simulation with agents moving on a grid."""
    
    def __init__(self, grid: Grid, arrival_rate: float = 2.0, 
                 dwell_time_range: Tuple[float, float] = (60, 120),
                 sink_time_range: Tuple[float, float] = (5, 15),
                 seconds_per_tick: float = 1.0,
                 p_male_urinal: float = 0.85):
        self.grid = grid
        self.arrival_rate = arrival_rate  # per minute
        self.dwell_time_range = dwell_time_range
        self.sink_time_range = sink_time_range
        self.seconds_per_tick = seconds_per_tick
        self.p_male_urinal = p_male_urinal
        
        self.people: List[Person] = []
        self.next_person_id = 1
        self.sim_time = 0.0
        
        # Stats
        self.served_count = 0
        self.total_time_in_system = 0.0
        self.female_count = 0
        self.male_count = 0
        self.female_time = 0.0
        self.male_time = 0.0
        
        self.rng = np.random.default_rng(42)
    
    def reset(self):
        self.people = []
        self.next_person_id = 1
        self.sim_time = 0.0
        self.served_count = 0
        for s in self.grid.stalls:
            s.occupant_id = None
            s.occupied_until = 0
        for s in self.grid.sinks:
            s.occupant_id = None
            s.occupied_until = 0
    
    def step(self):
        """One simulation tick."""
        self.sim_time += self.seconds_per_tick
        
        # Arrivals (Poisson)
        lambda_per_sec = self.arrival_rate / 60
        n_arrivals = self.rng.poisson(lambda_per_sec * self.seconds_per_tick)
        for _ in range(n_arrivals):
            self._spawn_person()
        
        # Update all agents
        for p in self.people:
            if p.state != PersonState.DONE:
                self._update_person(p)
        
        # Maintain queues
        self._maintain_queues()
    
    def _spawn_person(self):
        gender = 'F' if self.rng.random() < 0.5 else 'M'
        
        # Find entrance
        if gender == 'F' and self.grid.entrance_women:
            entrance = self.grid.entrance_women
        elif gender == 'M' and self.grid.entrance_men:
            entrance = self.grid.entrance_men
        elif self.grid.entrance_cell:
            entrance = self.grid.entrance_cell
        else:
            return
        
        dwell = self.rng.uniform(*self.dwell_time_range)
        sink = self.rng.uniform(*self.sink_time_range)
        
        p = Person(
            id=self.next_person_id,
            col=entrance[0], row=entrance[1],
            gender=gender,
            dwell_time=dwell,
            sink_time=sink,
            time_entered=self.sim_time
        )
        self.next_person_id += 1
        self.people.append(p)
    
    def _update_person(self, p: Person):
        if p.state == PersonState.WALKING_TO_QUEUE:
            self._walk_to_queue(p)
        elif p.state == PersonState.IN_QUEUE:
            self._in_queue(p)
        elif p.state == PersonState.WALKING_TO_STALL:
            self._walk_to_stall(p)
        elif p.state == PersonState.IN_STALL:
            self._in_stall(p)
        elif p.state == PersonState.WALKING_TO_SINK:
            self._walk_to_sink(p)
        elif p.state == PersonState.AT_SINK:
            self._at_sink(p)
        elif p.state == PersonState.EXITING:
            self._exiting(p)
    
    def _get_queue_cells(self, gender: str) -> List[QueueCell]:
        if self.grid.queue_cells_shared:
            return self.grid.queue_cells_shared
        return self.grid.queue_cells_women if gender == 'F' else self.grid.queue_cells_men
    
    def _walk_to_queue(self, p: Person):
        if p.target_queue_idx is None:
            queue = self._get_queue_cells(p.gender)
            occupied = {pp.target_queue_idx for pp in self.people 
                       if pp.gender == p.gender and pp.state != PersonState.DONE 
                       and pp.target_queue_idx is not None}
            for i in range(len(queue) - 1, -1, -1):
                if i not in occupied:
                    p.target_queue_idx = i
                    break
            if p.target_queue_idx is None:
                p.target_queue_idx = len(queue) - 1
        
        queue = self._get_queue_cells(p.gender)
        if p.target_queue_idx >= len(queue):
            p.state = PersonState.EXITING
            return
        
        target = queue[p.target_queue_idx]
        self._step_toward(p, target.col, target.row)
        
        if p.col == target.col and p.row == target.row:
            p.state = PersonState.IN_QUEUE
    
    def _in_queue(self, p: Person):
        # Stay in queue position
        queue = self._get_queue_cells(p.gender)
        if p.target_queue_idx is not None and p.target_queue_idx < len(queue):
            target = queue[p.target_queue_idx]
            p.col, p.row = target.col, target.row
        
        # Only front of queue can get a stall
        if p.target_queue_idx == 0:
            stall = self._find_available_stall(p.gender)
            if stall:
                stall.occupant_id = p.id
                p.target_stall = stall
                p.state = PersonState.WALKING_TO_STALL
                p.target_queue_idx = None
                self._shift_queue(p.gender)
    
    def _find_available_stall(self, gender: str) -> Optional[Stall]:
        available = [s for s in self.grid.stalls
                     if s.occupant_id is None and s.occupied_until <= self.sim_time]

        # Filter by gender compatibility
        if not self.grid.queue_cells_shared:  # Gender-segregated
            available = [s for s in available if s.gender_allowed == gender]

        # Men prefer urinals
        if gender == 'M' and self.rng.random() < self.p_male_urinal:
            urinals = [s for s in available if s.stall_type == 'urinal']
            if urinals:
                return urinals[0]

        # Filter out urinals for final selection if not chosen above
        non_urinals = [s for s in available if s.stall_type != 'urinal']
        return non_urinals[0] if non_urinals else (available[0] if available else None)
    
    def _shift_queue(self, gender: str):
        """Shift queue forward when front person leaves."""
        for pp in self.people:
            if (pp.gender == gender and 
                pp.state == PersonState.IN_QUEUE and 
                pp.target_queue_idx is not None and 
                pp.target_queue_idx > 0):
                pp.target_queue_idx -= 1

        # Try to serve new front person
        front_person = next((p for p in self.people 
                            if p.gender == gender 
                            and p.state == PersonState.IN_QUEUE 
                            and p.target_queue_idx == 0), None)
    
        if front_person:
            stall = self._find_available_stall(gender)
            if stall:
                stall.occupant_id = front_person.id
                front_person.target_stall = stall
                front_person.state = PersonState.WALKING_TO_STALL
    
    def _walk_to_stall(self, p: Person):
        if not p.target_stall:
            p.state = PersonState.WALKING_TO_SINK
            return

        target_col, target_row = p.target_stall.entrance_col, p.target_stall.entrance_row
        self._step_toward(p, target_col, target_row)

        if p.col == target_col and p.row == target_row:
            p.col, p.row = p.target_stall.col, p.target_stall.row
            p.state = PersonState.IN_STALL
            p.time_entered_stall = self.sim_time
            p.target_queue_idx = None
    
    def _in_stall(self, p: Person):
        if p.time_entered_stall and self.sim_time - p.time_entered_stall >= p.dwell_time:
            if p.target_stall:
                p.target_stall.occupant_id = None
                p.target_stall.occupied_until = 0
                p.target_stall = None
            p.state = PersonState.WALKING_TO_SINK
    
    def _walk_to_sink(self, p: Person):
        if not p.target_sink:
            available = [s for s in self.grid.sinks if s.occupant_id is None]
            if available:
                p.target_sink = available[0]
                p.target_sink.occupant_id = p.id
            else:
                p.state = PersonState.EXITING
                return
        
        target_col, target_row = p.target_sink.entrance_col, p.target_sink.entrance_row
        self._step_toward(p, target_col, target_row)
        
        if p.col == target_col and p.row == target_row:
            p.col, p.row = p.target_sink.col, p.target_sink.row
            p.state = PersonState.AT_SINK
            p.time_entered_sink = self.sim_time
    
    def _at_sink(self, p: Person):
        if p.time_entered_sink and self.sim_time - p.time_entered_sink >= p.sink_time:
            if p.target_sink:
                p.target_sink.occupant_id = None
                p.target_sink = None
            p.state = PersonState.EXITING
    
    def _exiting(self, p: Person):
        if not self.grid.exit_cell:
            p.state = PersonState.DONE
            return
        
        self._step_toward(p, self.grid.exit_cell[0], self.grid.exit_cell[1])
        
        if p.col == self.grid.exit_cell[0] and p.row == self.grid.exit_cell[1]:
            p.state = PersonState.DONE
            total_time = self.sim_time - p.time_entered
            self.served_count += 1
            self.total_time_in_system += total_time
            if p.gender == 'F':
                self.female_count += 1
                self.female_time += total_time
            else:
                self.male_count += 1
                self.male_time += total_time
    
    def _step_toward(self, p: Person, target_col: int, target_row: int):
        def is_walkable(col, row):
            if col < 0 or col >= self.grid.cols or row < 0 or row >= self.grid.rows:
                return False
            cell = self.grid.get_cell(row, col)
            if cell == CellType.WALL:
                return False
            if cell in [CellType.W_STALL, CellType.M_STALL, CellType.URINAL, CellType.SINK]:
                return (p.col == col and p.row == row)
            # Check collision with other people
            for other in self.people:
                if other != p and other.col == col and other.row == row and other.state != PersonState.DONE:
                    return False
            return True
        
        next_step = find_next_step(p.col, p.row, target_col, target_row,
                                   is_walkable, self.grid.cols, self.grid.rows)
        if next_step:
            p.col, p.row = next_step
    
    def _maintain_queues(self):
        for gender in ['F', 'M']:
            queue = self._get_queue_cells(gender)
            if not queue:
                continue
            
            occupied = {}
            for pp in self.people:
                if pp.gender == gender and pp.state == PersonState.IN_QUEUE:
                    if pp.target_queue_idx is not None:
                        occupied[pp.target_queue_idx] = pp
            
            # Fill gaps
            for i in range(len(queue)):
                if i not in occupied:
                    for j in range(i + 1, len(queue)):
                        if j in occupied:
                            occupied[j].target_queue_idx = i
                            break
```

## 6. Visualization Function

Draw the grid and agents using matplotlib.


```python
def draw_simulation(sim: CASimulation, ax=None, show_stats=True):
    """Draw the current state of the simulation."""
    if ax is None:
        fig, ax = plt.subplots(figsize=(14, 10))
    else:
        ax.clear()
    
    grid = sim.grid
    
    # Draw cells
    for row in range(grid.rows):
        for col in range(grid.cols):
            cell_type = grid.get_cell(row, col)
            color = CELL_COLORS.get(cell_type, '#ffffff')
            rect = plt.Rectangle((col, grid.rows - 1 - row), 1, 1, 
                                  facecolor=color, edgecolor='#cccccc', linewidth=0.5)
            ax.add_patch(rect)
            
            # Draw fixture symbols
            cx, cy = col + 0.5, grid.rows - 1 - row + 0.5
            
            if cell_type == CellType.W_STALL:
                ax.text(cx, cy, 'üö∫', ha='center', va='center', fontsize=12)
            elif cell_type == CellType.M_STALL:
                ax.text(cx, cy, 'üöπ', ha='center', va='center', fontsize=12)
            elif cell_type == CellType.URINAL:
                ax.text(cx, cy, 'U', ha='center', va='center', fontsize=10, fontweight='bold', color='white')
            elif cell_type == CellType.SINK:
                ax.text(cx, cy, 'üö∞', ha='center', va='center', fontsize=10)
            elif cell_type == CellType.ENTRANCE:
                ax.text(cx, cy, '‚Üí', ha='center', va='center', fontsize=14, color='green', fontweight='bold')
            elif cell_type == CellType.EXIT:
                ax.text(cx, cy, '‚Üê', ha='center', va='center', fontsize=14, color='red', fontweight='bold')
    
    # Draw occupancy indicators on fixtures
    for stall in grid.stalls:
        cx, cy = stall.col + 0.85, grid.rows - 1 - stall.row + 0.85
        color = 'red' if stall.occupant_id else 'green'
        circle = plt.Circle((cx, cy), 0.12, color=color, zorder=10)
        ax.add_patch(circle)
    
    for sink in grid.sinks:
        cx, cy = sink.col + 0.85, grid.rows - 1 - sink.row + 0.85
        color = 'red' if sink.occupant_id else 'green'
        circle = plt.Circle((cx, cy), 0.12, color=color, zorder=10)
        ax.add_patch(circle)
    
    # Draw people
    for p in sim.people:
        if p.state == PersonState.DONE:
            continue
        if p.state in [PersonState.IN_STALL, PersonState.AT_SINK]:
            continue  # Don't show people inside fixtures
        
        cx = p.col + 0.5
        cy = grid.rows - 1 - p.row + 0.5
        
        # Gender colors
        color = '#e91e63' if p.gender == 'F' else '#2196f3'
        circle = plt.Circle((cx, cy), 0.35, color=color, zorder=20)
        ax.add_patch(circle)
        
        # White border
        circle_border = plt.Circle((cx, cy), 0.35, fill=False, color='white', linewidth=2, zorder=21)
        ax.add_patch(circle_border)
        
        # State indicator (small dot)
        state_colors = {
            PersonState.WALKING_TO_QUEUE: '#ffeb3b',
            PersonState.IN_QUEUE: '#ff9800',
            PersonState.WALKING_TO_STALL: '#4caf50',
            PersonState.WALKING_TO_SINK: '#00bcd4',
            PersonState.EXITING: '#607d8b',
        }
        sc = state_colors.get(p.state, '#000000')
        small_dot = plt.Circle((cx, cy + 0.2), 0.1, color=sc, zorder=22)
        ax.add_patch(small_dot)
    
    ax.set_xlim(0, grid.cols)
    ax.set_ylim(0, grid.rows)
    ax.set_aspect('equal')
    ax.axis('off')
    
    # Stats text
    if show_stats:
        active = len([p for p in sim.people if p.state != PersonState.DONE])
        in_queue = len([p for p in sim.people if p.state == PersonState.IN_QUEUE])
        in_stall = len([p for p in sim.people if p.state == PersonState.IN_STALL])
        at_sink = len([p for p in sim.people if p.state == PersonState.AT_SINK])
        
        stats_text = f"Time: {sim.sim_time:.0f}s | Active: {active} | Queue: {in_queue} | Stall: {in_stall} | Sink: {at_sink} | Served: {sim.served_count}"
        ax.set_title(stats_text, fontsize=12, fontweight='bold')
    
    return ax
```

## 7. Create a Sample Layout

Build a gender-segregated bathroom layout.


```python
def build_gender_segregated_layout(n_stalls_female=3, n_stalls_male=2, n_urinals=2, n_sinks=2):
    """Build a gender-segregated bathroom layout."""
    cols = 24
    rows = 14
    grid = Grid(cols, rows)
    grid.draw_border()
    
    mid_col = cols // 2
    
    # Vertical divider
    for r in range(1, rows - 1):
        grid.set_cell(r, mid_col, CellType.WALL)
    
    # WOMEN'S SIDE (left)
    # Stalls at top
    for i in range(n_stalls_female):
        grid.add_stall(col=3 + i * 2, row=2, gender='F', stall_type='stall')
    
    # Sinks
    for i in range(n_sinks // 2):
        grid.add_sink(col=4 + i * 2, row=5)
    
    # Queue cells (row 9)
    for c in range(2, mid_col):
        grid.add_queue_cell(col=c, row=9, gender='F')
    
    # Entrance
    grid.set_entrance(col=5, row=rows - 2, gender='F')
    
    # MEN'S SIDE (right)
    # Stalls
    for i in range(n_stalls_male):
        grid.add_stall(col=mid_col + 3 + i * 2, row=2, gender='M', stall_type='stall')
    
    # Urinals
    for i in range(n_urinals):
        grid.add_stall(col=mid_col + 3 + n_stalls_male * 2 + i, row=2, gender='M', stall_type='urinal')
    
    # Sinks
    for i in range(n_sinks // 2):
        grid.add_sink(col=mid_col + 4 + i * 2, row=5)
    
    # Queue cells (row 9)
    for c in range(mid_col + 1, cols - 2):
        grid.add_queue_cell(col=c, row=9, gender='M')
    
    # Entrance
    grid.set_entrance(col=mid_col + 5, row=rows - 2, gender='M')
    
    # Shared exit
    grid.set_exit(col=mid_col, row=rows - 2)
    
    return grid


def build_all_gender_layout(n_stalls=6, n_sinks=2):
    """Build an all-gender bathroom layout."""
    cols = 20
    rows = 14
    grid = Grid(cols, rows)
    grid.draw_border()
    
    # Stalls at top (all shared)
    start_col = (cols - n_stalls * 2) // 2 + 1
    for i in range(n_stalls):
        grid.add_stall(col=start_col + i * 2, row=2, gender='F', stall_type='stall')
        # Make them shared
        grid.stalls[-1].gender_allowed = 'both'
    
    # Sinks
    sink_start = (cols - n_sinks * 2) // 2
    for i in range(n_sinks):
        grid.add_sink(col=sink_start + i * 2, row=5)
    
    # Single shared queue (row 9)
    for c in range(3, cols - 3):
        grid.add_queue_cell(col=c, row=9, gender='shared')
    
    # Shared entrance
    grid.set_entrance(col=cols // 2, row=rows - 2)
    
    # Exit
    grid.set_exit(col=cols - 3, row=rows - 2)
    
    return grid


# Create the layouts
print("Creating layouts...")
grid_segregated = build_gender_segregated_layout(n_stalls_female=3, n_stalls_male=2, n_urinals=2, n_sinks=2)
grid_allgender = build_all_gender_layout(n_stalls=6, n_sinks=2)

print(f"Gender-segregated: {len(grid_segregated.stalls)} fixtures, {len(grid_segregated.sinks)} sinks")
print(f"All-gender: {len(grid_allgender.stalls)} fixtures, {len(grid_allgender.sinks)} sinks")
```

## 8. Run and Visualize the Simulation

Run the simulation and see agents moving on the grid!


```python
# Create simulation with the gender-segregated layout
sim = CASimulation(
    grid=grid_segregated,
    arrival_rate=3.0,  # 3 people per minute
    dwell_time_range=(30, 90),  # 30-90 seconds in stall
    sink_time_range=(5, 15),     # 5-15 seconds at sink
    seconds_per_tick=2.0         # 2 seconds per step (faster for visualization)
)

# Show initial layout
fig, ax = plt.subplots(figsize=(14, 8))
draw_simulation(sim, ax)
plt.suptitle("Initial Layout - Gender Segregated Bathroom", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("\nLegend:")
print("  üü£ Pink circles = Women")
print("  üîµ Blue circles = Men")
print("  üü¢ Green dots on fixtures = Available")
print("  üî¥ Red dots on fixtures = Occupied")
print("  Pink cells = Women's queue | Blue cells = Men's queue")
```

```python
# Run simulation step by step with visualization
# This shows snapshots every N steps

def run_visual_simulation(sim, n_steps=100, snapshot_interval=10):
    """Run simulation and show snapshots at intervals."""
    sim.reset()
    
    snapshots = []
    
    for step in range(n_steps):
        sim.step()
        
        if step % snapshot_interval == 0 or step == n_steps - 1:
            # Capture current state
            people_snapshot = [(p.col, p.row, p.gender, p.state) 
                              for p in sim.people if p.state != PersonState.DONE]
            stall_occupied = [s.occupant_id is not None for s in sim.grid.stalls]
            sink_occupied = [s.occupant_id is not None for s in sim.grid.sinks]
            
            snapshots.append({
                'step': step,
                'time': sim.sim_time,
                'people': people_snapshot,
                'stalls': stall_occupied,
                'sinks': sink_occupied,
                'served': sim.served_count,
                'active': len(people_snapshot)
            })
    
    return snapshots

# Run and capture snapshots
print("Running simulation...")
snapshots = run_visual_simulation(sim, n_steps=150, snapshot_interval=15)
print(f"Captured {len(snapshots)} snapshots")

# Display snapshots in a grid
n_cols = 3
n_rows = (len(snapshots) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 5 * n_rows))
axes = axes.flatten()

for idx, snap in enumerate(snapshots):
    ax = axes[idx]
    
    # Reset simulation state to this snapshot for drawing
    # (We'll manually draw since we only have snapshot data)
    
    grid = sim.grid
    
    # Draw cells
    for row in range(grid.rows):
        for col in range(grid.cols):
            cell_type = grid.get_cell(row, col)
            color = CELL_COLORS.get(cell_type, '#ffffff')
            rect = plt.Rectangle((col, grid.rows - 1 - row), 1, 1, 
                                  facecolor=color, edgecolor='#cccccc', linewidth=0.3)
            ax.add_patch(rect)
    
    # Draw fixture occupancy
    for i, stall in enumerate(grid.stalls):
        cx, cy = stall.col + 0.85, grid.rows - 1 - stall.row + 0.85
        color = 'red' if snap['stalls'][i] else 'green'
        circle = plt.Circle((cx, cy), 0.15, color=color, zorder=10)
        ax.add_patch(circle)
    
    for i, sink in enumerate(grid.sinks):
        cx, cy = sink.col + 0.85, grid.rows - 1 - sink.row + 0.85
        color = 'red' if snap['sinks'][i] else 'green'
        circle = plt.Circle((cx, cy), 0.15, color=color, zorder=10)
        ax.add_patch(circle)
    
    # Draw people from snapshot
    for (col, row, gender, state) in snap['people']:
        if state in [PersonState.IN_STALL, PersonState.AT_SINK]:
            continue
        
        cx = col + 0.5
        cy = grid.rows - 1 - row + 0.5
        
        color = '#e91e63' if gender == 'F' else '#2196f3'
        circle = plt.Circle((cx, cy), 0.35, color=color, zorder=20)
        ax.add_patch(circle)
        circle_border = plt.Circle((cx, cy), 0.35, fill=False, color='white', linewidth=1.5, zorder=21)
        ax.add_patch(circle_border)
    
    ax.set_xlim(0, grid.cols)
    ax.set_ylim(0, grid.rows)
    ax.set_aspect('equal')
    ax.axis('off')
    ax.set_title(f"t={snap['time']:.0f}s | Active:{snap['active']} | Served:{snap['served']}", fontsize=10)

# Hide empty subplots
for idx in range(len(snapshots), len(axes)):
    axes[idx].axis('off')

plt.suptitle("Simulation Snapshots - Watch People Move Through the System", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 9. Compare Layouts: Gender-Segregated vs All-Gender

Run both layouts and compare queue lengths and wait times.


```python
def run_and_collect_stats(grid, name, n_steps=300, arrival_rate=4.0):
    """Run a simulation and collect statistics."""
    sim = CASimulation(
        grid=grid,
        arrival_rate=arrival_rate,
        dwell_time_range=(40, 100),
        sink_time_range=(5, 15),
        seconds_per_tick=2.0
    )
    
    queue_lengths = []
    female_in_queue = []
    male_in_queue = []
    
    for step in range(n_steps):
        sim.step()
        
        # Track queue lengths
        in_queue = [p for p in sim.people if p.state == PersonState.IN_QUEUE]
        queue_lengths.append(len(in_queue))
        female_in_queue.append(len([p for p in in_queue if p.gender == 'F']))
        male_in_queue.append(len([p for p in in_queue if p.gender == 'M']))
    
    # Final stats
    avg_time_f = sim.female_time / sim.female_count if sim.female_count > 0 else 0
    avg_time_m = sim.male_time / sim.male_count if sim.male_count > 0 else 0
    
    return {
        'name': name,
        'queue_lengths': queue_lengths,
        'female_queue': female_in_queue,
        'male_queue': male_in_queue,
        'served': sim.served_count,
        'female_count': sim.female_count,
        'male_count': sim.male_count,
        'avg_time_female': avg_time_f,
        'avg_time_male': avg_time_m,
        'wait_gap': avg_time_f - avg_time_m
    }

# Run both scenarios
print("Running Gender-Segregated scenario...")
stats_seg = run_and_collect_stats(grid_segregated, "Gender-Segregated", n_steps=300, arrival_rate=4.0)

print("Running All-Gender scenario...")
stats_all = run_and_collect_stats(grid_allgender, "All-Gender", n_steps=300, arrival_rate=4.0)

# Plot comparison
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Queue lengths over time
ax = axes[0, 0]
ax.plot(stats_seg['queue_lengths'], label='Segregated', alpha=0.7)
ax.plot(stats_all['queue_lengths'], label='All-Gender', alpha=0.7)
ax.set_xlabel('Time Step')
ax.set_ylabel('Total Queue Length')
ax.set_title('Queue Length Over Time')
ax.legend()
ax.grid(True, alpha=0.3)

# Queue by gender (segregated)
ax = axes[0, 1]
ax.plot(stats_seg['female_queue'], label='Women (Segregated)', color='#e91e63', alpha=0.7)
ax.plot(stats_seg['male_queue'], label='Men (Segregated)', color='#2196f3', alpha=0.7)
ax.set_xlabel('Time Step')
ax.set_ylabel('Queue Length')
ax.set_title('Gender-Segregated: Queue by Gender')
ax.legend()
ax.grid(True, alpha=0.3)

# Average time in system
ax = axes[1, 0]
scenarios = ['Segregated', 'All-Gender']
female_times = [stats_seg['avg_time_female'], stats_all['avg_time_female']]
male_times = [stats_seg['avg_time_male'], stats_all['avg_time_male']]

x = np.arange(len(scenarios))
width = 0.35
ax.bar(x - width/2, female_times, width, label='Women', color='#e91e63')
ax.bar(x + width/2, male_times, width, label='Men', color='#2196f3')
ax.set_ylabel('Avg Time in System (seconds)')
ax.set_title('Average Time in System by Gender')
ax.set_xticks(x)
ax.set_xticklabels(scenarios)
ax.legend()
ax.grid(True, alpha=0.3, axis='y')

# Wait gap
ax = axes[1, 1]
gaps = [stats_seg['wait_gap'], stats_all['wait_gap']]
colors = ['#ff4444' if g > 0 else '#44aa44' for g in gaps]
ax.bar(scenarios, gaps, color=colors)
ax.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
ax.set_ylabel('Wait Gap (Women - Men, seconds)')
ax.set_title('Wait Gap (positive = women wait longer)')
ax.grid(True, alpha=0.3, axis='y')

plt.suptitle('Comparison: Gender-Segregated vs All-Gender Layout', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# Print summary
print("\n" + "="*60)
print("SUMMARY")
print("="*60)
for stats in [stats_seg, stats_all]:
    print(f"\n{stats['name']}:")
    print(f"  Served: {stats['served']} ({stats['female_count']} F, {stats['male_count']} M)")
    print(f"  Avg time (Women): {stats['avg_time_female']:.1f}s")
    print(f"  Avg time (Men): {stats['avg_time_male']:.1f}s")
    print(f"  Wait Gap: {stats['wait_gap']:.1f}s {'(women wait longer)' if stats['wait_gap'] > 0 else '(men wait longer)'}")
```
